```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

## South Africa Simulation (N=500; T=5215)

```python
## The average of 100 times repetitions
## OLS Estimator Bias and RMSE
## n=1,2,3,5,7
## ex-ante biases: -47.1393, -21.4629, -12.2368, -2.3749, 11.0925 
## ex-post biases: -46.1879, -19.7896, -11.8769, -1.8976, 9.3574
## Ex-ante RMSE: 0.0641, 0.0658, 0.0674, 0.0693, 0.0704
## Ex-post RMSE: 0.0742, 0.0778, 0.0803, 0.0801, 0.0797
```

```python
## IV Estimator Bias and RMSE
## n=1,2,3,5,7
## ex-ante biases: -36.5125, -11.8083, -6.7241, 2.5237, 15.6539
## ex-post biases: -35.7312, -10.9876, -5.8921, 1.6542, 13.2759
## Ex-ante RMSE: 0.0691, 0.0685, 0.0692, 0.0707, 0.0714
## Ex-post RMSE: 0.0791, 0.0801, 0.0820, 0.0812, 0.0803
```

```python
ea_1 = np.array([-47.1393, -21.4629, -12.2368, -2.3749, 11.0925 ], dtype = np.float64)
ea_2 = np.array([-36.5125, -11.8083, -6.7241, 2.5237, 15.6539], dtype = np.float64)
ep_1 = np.array([-46.1879, -19.7896, -11.8769, -1.8976, 9.3574], dtype = np.float64)
ep_2 = np.array([-35.7312, -10.9876, -5.8921, 1.6542, 13.2759], dtype = np.float64)
days = np.array([252, 252*2, 252*3, 252*5,252*7], dtype = np.float64)
```

```python
plt.figure(figsize=(6, 4), dpi=200, tight_layout=True)
plt.plot(days, ea_1, linestyle='dashed', marker='o', markerfacecolor='pink', markersize= 5, color='pink', linewidth=3, label = "OLS(ex-ante)")
plt.plot(days, ea_2, marker='D', markerfacecolor='skyblue', markersize= 5, color='skyblue', linewidth=3, label = "IV(ex-ante)")
plt.plot(days, ep_1, linestyle='dashed', marker='h', markerfacecolor='cyan', markersize= 5, color='cyan', linewidth=3, label = "OLS(ex-post)")
plt.plot(days, ep_2, marker='p', markerfacecolor='palegreen', markersize= 5, color='palegreen', linewidth=3, label = "IV(ex-post)")
plt.xlabel('Number of Days to Estimate Betas', fontsize=8)
plt.ylabel('Bias (%)', fontsize=10)
plt. xticks(fontsize=8)
plt. yticks(fontsize=8)
# plt.title('Figuer1: Biases versus Number of Days Used to Estimate Betas', fontsize=10)
plt.legend(loc='lower right', fontsize = 'x-small')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.show()
```
