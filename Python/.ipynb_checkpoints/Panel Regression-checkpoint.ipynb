```python
def nansum(vec, **kwargs): # vec, axis=1
    mx = np.isnan(vec).all(**kwargs)
    res = np.nansum(vec, **kwargs)
    res[mx] = np.nan
    return res
```

```python
"""
Input Explanation:
df_beta is the dataframe of stock betas obtained by running the first stage of BJS regression  # mat(N*K)
df_f is the dataframe of factor realizations # mat(T*K)
df_lam is the dataframe of factor risk premiums obtained by running the second stage of BJS regression # mat(T*K)
df_res is the dataframe of residuals obtained by running the first stage of BJS regression # mat(T*N)
"""

def Shanken_BJS_tstats(df_betas, df_f, df_lambdas, df_res):
    beta = df_betas
    lam = df_lambdas.mean()
    f = np.mat(df_f.T)
    res = df_res
    
    beta_mul = np.dot(np.linalg.inv(np.dot(beta.T,beta)),beta.T) #mat(K*N)
    beta_rev = np.dot(beta,np.linalg.inv(np.dot(beta.T,beta))) #mat(N*K)
    cov_res = res.cov() #mat(N*N)
    beta_mul_res = np.dot(np.dot(beta_mul, cov_res),beta_rev) #mat(K*K)
    cov_f = np.mat(np.cov(f)) #mat(K*K)
    c = pd.DataFrame(1+np.dot(np.dot(lam,np.linalg.inv(cov_f)),lam.T)) #c is a coefficient 
    
    adj_cov = (1/len(df_f))*(beta_mul_res*c.values+cov_f) #mat(K*K)
    se = np.sqrt(np.diagonal(adj_cov))
    t_stats = lam.values/se
    return t_stats
```

```python
def replacer(l,target_val=0,replace_val=np.NaN,repeat_max=10):
    #repeat_max = n (consecutive 0s over 10 times will be replaced with NAs)
    counter = 0
    new_l = []
    for e in l:
        if e == target_val: counter += 1
        else:
            counter = 0

        if counter > repeat_max:
            new_l.append(replace_val)
        else:
            new_l.append(e)

    return new_l
```

```python
def froll_sum(x): 
    # Convert daily return to monthly return, use fix_rolling_sum function
    tmp = []
    for i in range(int(len(x)/21)): # 21 represents the trading days for one month
        res = np.nansum(x[i*21:(i+1)*21])
        tmp.append(res)
    return tmp
```

```python
## Newey-West Adjustment For a Time-Series Data, e.g. Lambdas (factor returns)
def NW_ttest(ret,h0):
  T = len(ret)
  lag = int(4*(T/100)**(2/9))
  vv = np.var(ret)

  for l in range(0,lag):
      cc=np.cov(ret[0:(len(ret)-l)].T,ret[l:len(ret)].T)[0,1]
      vv=vv+2*(1-(l+1)/lag)*cc
      t=(np.mean(ret)-h0)/np.sqrt(vv)*np.sqrt(T)
  
  return t
```

```python
holding = 1
rollingW = holding*n*2
lens = len(m_df_MR)
periods = int((lens - rollingW) / holding) - 1
print(periods, rollingW)
```

# Panel Regression

```python
# Determine how many factors are involved in the regression
ran = range(1,7)
print(ran)
```

```python
from linearmodels import PooledOLS
from linearmodels.panel import PanelOLS
exog = sm.add_constant(panel_trnct.iloc[:,ran])
endog = panel_trnct["Return"]
mod = PanelOLS(endog, exog)
panelOLS_res = mod.fit(cov_type='clustered', cluster_entity=True)
```

```python
panelOLS_res.summary
```

```python
# Store values for checking homoskedasticity graphically
fittedvals_panel_OLS = panelOLS_res.predict().fitted_values
residuals_panel_OLS = panelOLS_res.resids
```

```python
# Homoskedasticity
## Residuals-Plot for growing Variance Detection
fig, ax = plt.subplots()
ax.scatter(fittedvals_panel_OLS, residuals_panel_OLS, color = "blue")
ax.axhline(0, color = 'r', ls = '--')
ax.set_xlabel("Predicted Values", fontsize = 15)
ax.set_ylabel("Residuals", fontsize = 15)
ax.set_title("Homoskedasticity Test", fontsize = 30)
plt.show()
```

```python
## White-Test
from statsmodels.stats.diagnostic import het_white, het_breuschpagan
panel_OLS_dataset = pd.concat([panel_trnct, residuals_panel_OLS], axis=1)
panel_OLS_dataset = panel_OLS_dataset.drop(["Period"], axis = 1)
exog = sm.add_constant(panel_trnct.iloc[:,ran])
white_test_results = het_white(panel_OLS_dataset["residual"], exog)
labels = ["LM-Stat", "LM p-val", "F-Stat", "F p-val"] 
print(dict(zip(labels, white_test_results)))
## Breusch-Pagan-Test
breusch_pagan_test_results = het_breuschpagan(panel_OLS_dataset["residual"], exog)
labels = ["LM-Stat", "LM p-val", "F-Stat", "F p-val"] 
print(dict(zip(labels, breusch_pagan_test_results)))
```

**Heteroskedasticity: In simple terms, if p < 0.05, then heteroskedasticity is indicated.**

```python
## Non-Autocorrelation
## Durbin-Watson-Test
from statsmodels.stats.stattools import durbin_watson

durbin_watson_test_results = durbin_watson(panel_OLS_dataset["residual"]) 
print(durbin_watson_test_results)
```

**Auto-Correlation: The Durbin-Watson-Test will have one output between 0  4. The mean (= 2) would indicate that there is no autocorrelation identified, 0  2 means positive autocorrelation (the nearer to zero the higher the correlation), and 2  4 means negative autocorrelation (the nearer to four the higher the correlation).**

To sum up, assumption 3a and 3b are violated in this sample, so it seems that a FE-/RE-model will be more suitable.

```python
# FE und RE model
from linearmodels import PanelOLS
from linearmodels import RandomEffects
exog = sm.add_constant(panel_trnct.iloc[:,ran])
endog = panel_trnct["Return"]
# random effects model
model_re = RandomEffects(endog, exog) 
re_res = model_re.fit() 
# fixed effects model
model_fe = PanelOLS(endog, exog, entity_effects = True) # if entity_effects is true, then panelOLS would become fixed effect model
fe_res = model_fe.fit() 
#print results
print(re_res)
print(fe_res)
```

```python
import numpy.linalg as la
from scipy import stats
import numpy as np

def hausman(fe, re):
    b = fe.params
    B = re.params
    v_b = fe.cov
    v_B = re.cov
    df = b[np.abs(b) < 1e8].size
    chi2 = np.dot((b - B).T, la.inv(v_b - v_B).dot(b - B)) 
 
    pval = stats.chi2.sf(chi2, df)
 
    return chi2, df, pval

hausman_results = hausman(fe_res, re_res) 

print("chi-Squared: " + str(hausman_results[0]))
print("degrees of freedom: " + str(hausman_results[1]))
print("p-Value: " + str(hausman_results[2]))
```

**Since the p-value is very small, the null hypothesis should be rejected. Accordingly, the FE-model seems to be the most suitable, because we clearly do have endogeneity in our model.**
